```python
import csv
```

```python
f = open('cig.csv' , 'rb')
reader = csv.reader(f)
tar = []
nicotine = []

for row in reader :
    a = row[0].replace('\t' , ' ' )
    b = a.split()
    c = b[1:]
    d = map(float , c)
    e = b [:1]
    e.extend(d)
    tar.append(e[1])
    nicotine.append(e[2])
print tar
print nicotine
f.close()
 
```

```python
from pylab import *
```

```python
scatter(nicotine , tar)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('cigarret`s tar & nicotine')
ax.set_xlabel(' cigarret`s tar(g)')
ax.set_ylabel('cigarret`s nicotine (cc)')
ax.scatter(nicotine , tar)
fig.savefig("cigarrets_with_titles.png")
```
